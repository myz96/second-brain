```python
import os
from langchain_community.document_loaders import UnstructuredMarkdownLoader

# Directory containing your Markdown files
directory = "data/"

# List to store all loaded documents
all_docs = []

# Iterate over all files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".md"):  # Check if the file is a Markdown file
        file_path = os.path.join(directory, filename)
        
        loader = UnstructuredMarkdownLoader(
            file_path,
            mode="single",
            strategy="fast",
        )
        
        docs = loader.load()
        all_docs.extend(docs)
        
        print(f"Loaded content from {filename}:")
        print(docs[0].page_content[:200])  # Print first 200 characters
        print("\n" + "="*50 + "\n")  # Separator between files

print(f"Total documents loaded: {len(all_docs)}")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    add_start_index=True,
)
all_splits = text_splitter.split_documents(all_docs)

len(all_splits)
```
